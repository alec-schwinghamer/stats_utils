```python
def false_positive_risk(alpha, power, success_rate):
    """Calculates the False Positive Risk (or probability that a statistically significant result is a false positive, i.e. the probability that the null 
    hypothesis is true when an experiment was statistically significant)"""
    pi = 1 - success_rate
    FPR = alpha*pi/(((alpha*pi)) + ((power)*(1-pi)))
    
    return FPR
```

```python
false_positive_risk(0.025, 0.5, 0.05)
```

```python
x = np.linspace(0.05, 0.5, 100)
```

```python
x
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, false_positive_risk(alpha=0.025, power=0.5, success_rate=x), color='blue', linewidth=3)
plt.plot(0.333, false_positive_risk(alpha=0.025, power=0.5, success_rate=0.333), 'go', markersize=12, label="If our success rate was that of Microsoft") 
plt.plot(0.15, false_positive_risk(alpha=0.025, power=0.5, success_rate=0.15), 'rD', markersize=12, label="If our success rate was that of Bing")
plt.plot(0.1, false_positive_risk(alpha=0.025, power=0.5, success_rate=0.1), 'mH', markersize=12, label="If our success rate was that of Google Ads or Netflix") 
plt.plot(0.08, false_positive_risk(alpha=0.025, power=0.5, success_rate=0.08), 'kv', markersize=12, label="If our success rate was that of Airbnb")
plt.title("False Positive Risk (FPR) as a function of experiment success rate at alpha=0.025, beta=0.2", fontsize=18)
plt.xlabel('Historical experiment success rate', fontsize=18)
plt.xticks(fontsize=14)
plt.ylabel("False Positive Risk", fontsize=18)
plt.yticks(fontsize=14)
plt.legend()
plt.tight_layout()
plt.savefig("FPR.png", facecolor='w')
```
